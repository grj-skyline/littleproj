```python
import matplotlib.pyplot as plt
import numpy as np 
from qutip import*
```

```python
H = tensor(sigmax(),qeye(2))+tensor(qeye(2),sigmax())+ 0.05*tensor(sigmax(),sigmax())
psi0 = tensor(basis(2,0),basis(2,1))

tlist = np.linspace(0,20,100)
H
```

```python
result = mesolve(H , psi0 ,tlist ,[],[tensor(sigmaz(),qeye(2)),tensor(qeye(2),sigmaz())])
```

```python
result.expect[0]
```

```python
fig,axes = plt.subplots(1,1)

axes.plot(tlist,result.expect[0])
axes.plot(tlist,result.expect[1])
```
