```python
import matplotlib.pyplot as plt
import numpy as np
from qutip import*
```


$$H = \frac{\sqrt{3}}{2} at(q_x \tau_z \sigma_x + q_y \sigma_y) + \frac{\Delta}{2}\sigma_z$$

```python
def H_0(Delta,k1,k2,t,tauz):
    
    
    b1 = 4*np.pi/3*np.array([1/2,np.sqrt(3)/2])
    b2 = 4*np.pi/3*np.array([-1/2,np.sqrt(3)/2])
    
    k = k1*b1 + k2*b2
    
    qx = np.inner(k,b1)
    qy = np.inner(k,b2)
    
    H = np.sqrt(3)/2*(qx*tauz*sigmax() + qy*sigmay()) + Delta/2*sigmaz()
    
    evals = H.eigenenergies()
    return (evals)
```

```python
fig, ax = plt.subplots()
step_k = 100
k = np.linspace(-1/10*np.pi,1/10*np.pi,step_k)
step_k_2 = 2
k_2 = np.linspace(-np.pi,np.pi,step_k_2)
for u in k_2:
    result = []
    for i in k :
        result.append(H_0(1,i,0,1,1))
    for j in range(2):
        ax.plot(k,[temp[j] for temp in result],'black')
        
```
