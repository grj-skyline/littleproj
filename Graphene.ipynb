```python
import matplotlib.pyplot as plt
import numpy as np
from qutip import*
```

```python
def Energy_k(w,k1,k2,phi,t):
    delta1 = np.array([0,1])
    #delta2 = np.array([np.sqrt(3)/2,-0.5])
    delta2 = np.array([np.sqrt(3)/2,-0.5])
    delta3 = np.array([-np.sqrt(3)/2,-0.5])
    
    g1 = np.array([1/2,np.sqrt(3)/2])
    g2 = np.array([-1/2,np.sqrt(3)/2])
    
    k = k1*g1 + k2*g2
    alpha1 = np.inner(k,delta1)
    alpha2 = np.inner(k,delta2)
    alpha3 = np.inner(k,delta3)
    H = Qobj([[w,-t*(np.exp(1j*alpha1) + np.exp(1j*alpha2) + np.exp(1j*alpha3))],
             [-t*(np.exp(-1j*alpha1)+np.exp(-1j*alpha2)+np.exp(-1j*alpha3)),-w]
             ])
    evals = H.eigenenergies()
    return(evals)
```

```python
fig, ax = plt.subplots()
step_k = 100
k = np.linspace(-np.pi,np.pi,step_k)
step_k_2 = 30
k_2 = np.linspace(-10*np.pi,10*np.pi,step_k_2)
for u in k_2:
    result = []
    for i in k :
        result.append(Energy_k(0,i,u,0,1))
    for j in range(2):
        ax.plot(k,[temp[j] for temp in result],'black')
```
