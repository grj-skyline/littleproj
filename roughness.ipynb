```python
import numpy as np
import numpy.matlib 
import random

```

```python
#orderï¼šk*l*m 100*100*25
```

```python
G = np.zeros((31,31,15))
for k in range (0,31):
    for l in range (0,31):
        for m in range(0,15):
            if k==0 and l==0 and m==0:
                continue
            else:
                G[k,l,m] = (2*m*m-k*k-l*l)/((k*k+l*l+m*m)**(5/2))
print(G)            
```

```python
p = np.matlib.zeros((30*30*15,1))
Eex = np.matlib.zeros((30*30*15,1))

alpha1 = 1
alpha2 = 1

for i in range (0, 30*30*15):
    if i < 30*30*10 :
        Eex[i] = 5*alpha1
    else:
        Eex[i] = 5*np.exp(0.1*(-int(i/900)+9))*alpha2
```

```python
print(Eex[900*10])
```

```python
S = np.zeros((30,30))
for i in range (0,30):
    for j in range(0,30):
        S[i,j] = 10
for i in range(0,1000):
    x = random.randint(0,29)
    y = random.randint(0,29)
    if S[x,y] < 15:
        S[x,y] = S[x,y]+1
#print(S)
```

```python
M = np.matlib.zeros((30*30*15,30*30*15))

for i in range(0,30*30*15):
    if i%1000 == 0:
        print(i)
        
        #print(M[i,j])
    for j in range(0,30*30*15):
        l1 = (i%900)%30
        k1 = int((i%900)/30)
        m1 = int(i/900)
        
        l2 = (j%900)%30
        k2 = int((j%900)/30)
        m2 = int(j/900)
        
        if i == j:
            M[i,j] = 1.00000
        else :
            M[i,j] = alpha1*G[abs(l1-l2),abs(k1-k2),abs(m1-m2)]
        
```

```python
p = np.linalg.solve(M,Eex)
```

```python
print(M)
```

```python
Polarizations = np.zeros(30*30*15)
for i in range(30*30*15):
    Polarizations[i] = p[i]

Polarizations = Polarizations.reshape((15,30,30))
```

```python
print(Polarizations)
```

```python

np.save(file="Polarizations.npy", arr=Polarizations)
```
